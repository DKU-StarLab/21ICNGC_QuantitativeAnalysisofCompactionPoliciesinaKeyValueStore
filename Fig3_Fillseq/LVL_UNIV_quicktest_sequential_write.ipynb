```python
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
import re 
import sys
import os
```

```python
def search(dirname, _pattern):
    stringbuilder=[]
    filenames = os.listdir(dirname) 
    for file in filenames:
        if _pattern.match(file) !=None :
            stringbuilder.append(str(file))
    return stringbuilder
```

```python
def inhosort(FILE):
    stringbuilder=[]
    for file in FILE:
        l = [int(re.search('\d+',re.search('k\d+',file).group(0)).group(0))
                ,int(re.search('\d+',re.search('v\d+',file).group(0)).group(0))
                 ,file
            , file.split('_')[2]]
        stringbuilder.append(l)
    stringbuilder.sort()
    return stringbuilder
```

```python
def inhosort_cache(FILE):
    stringbuilder=[]
    for file in FILE:
        l = [int(re.search('\d+',re.search('k\d+',file).group(0)).group(0))
             ,int(re.search('\d+',re.search('v\d+',file).group(0)).group(0))
             ,int(re.search('\d+',re.search('c\d+',file).group(0)).group(0))
             ,file
            , file.split('_')[2]]
        stringbuilder.append(l)
    stringbuilder.sort()
    return stringbuilder
```
